```python
import logging
def log_message(error_name,e):
    FORMAT = "%(asctime)-15s  %(levelname)s"
    logging.basicConfig(filename='log_message.log', level=logging.DEBUG, format=FORMAT)
    logging.exception(error_name+' is occurred')
    return None
```

```python
def sum(a):
    try:
        b = a + 6
    except TypeError as e:
        b = 0
        print('TypeError', e)
        log_message('TypeError', e)

    return b
```

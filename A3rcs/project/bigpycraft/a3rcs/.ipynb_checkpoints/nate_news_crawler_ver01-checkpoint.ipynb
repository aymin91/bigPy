```python
import requests

from bs4 import BeautifulSoup
import pandas as pd

import datetime
import re
```

```python
BASE_URL = 'https://news.nate.com/'

# 정치 / 경제 / 사회 param
PARAM = {'pol':'section?mid=n0200', 'eco':'section?mid=n0300', 'soc':'section?mid=n0400'}

# 기간 설정
DEFAULT_DATE = '2000'
```

#### step 1. 분야별 언론사 구하기

```python
def input_news_section():
    section = input('insert the section(pol/eco/soc) : ')

    if section == 'pol':
        sec_parm = PARAM['pol']
    elif section == 'eco':
        sec_parm = PARAM['eco']
    elif section == 'soc':
        sec_parm = PARAM['soc']
    else :
        print('please check the section')
    
    return sec_parm
```

```python
def find_company_info_tag(sec_parm):

    req = requests.get(BASE_URL+sec_parm)
    soup = BeautifulSoup(req.text, 'html.parser')

    company_info_tag = soup.find('ul',class_='f_clear')

    return company_info_tag
```

```python
def get_company_info_tag(company_info_tag):
    
    company_info = dict()
    
    com = company_info_tag.find_all('li')
    
    for c in com:
        company_info[c.text] = c.a['href'][29:31]
    
    return company_info
```

```python
def make_company_info():
    
    sec_parm = input_news_section()
    company_info_tag = find_company_info_tag(sec_parm)
    company_info = get_company_info_tag(company_info_tag)
    
    return company_info
    
```

```python
company_info = make_company_info()
company_info
```

#### step2. 날짜 생성

```python
def make_datetime():
    
    init_day = datetime.date(int(DEFAULT_DATE), 1, 1)
    end_day = datetime.date(int(str(datetime.datetime.now().year)), int(str(datetime.datetime.now().month)), int(str(datetime.datetime.now().day)))
    
    day = init_day
    date_range = []

    while day < end_day:
        date_range.append(day.strftime("%Y%m%d"))
        day += datetime.timedelta(days=1)
        
    return date_range
```

```python
date_range = make_datetime()
date_range
```
